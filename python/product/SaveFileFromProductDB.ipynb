```python
import pymysql
import pandas as pd
```

```python
# DB에 접속 
db = pymysql.connect(
    host = 'IP',
    port = 3306,
    user = 'eddi',
    passwd = '비밀번호',
    db = 'buy_idea_db'
)

cursor = db.cursor()

sql1 = """
    select * from product;
"""

sql2 = """
    select * from product_info;
"""

sql3 = """
    select * from product_image;
"""

cursor.execute(sql1)
cursor.execute(sql2)
cursor.execute(sql3)


rows = cursor.fetchall()

# DB데이터 조회하기
for row in rows:
    print(row)
print('파이썬을 활용한 db 테이블 조회 성공!')

# DB데이터 cvs파일로 저장
productData = pd.read_sql_query(sql1, db)
productInfoData = pd.read_sql_query(sql2, db)
productImageData = pd.read_sql_query(sql3, db)

productData.to_csv('product_data6.csv', index=False)
productInfoData.to_csv('product_info_data6.csv', index=False)
productImageData.to_csv('product_image_data6.csv', index=False)
print('DB 데이터 cvs파일 저장 성공!')
    
db.close()
```

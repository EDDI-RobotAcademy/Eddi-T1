```python
f = open('product_data.csv', 'r', encoding='UTF-8')
csv_reader = csv.reader(f, delimiter=',')

print(csv_reader)

cnt = 0

cursor = db.cursor()

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
    
    nickname = (row[1])
    price = (row[2])
    title = (row[3])

        
    sql = "INSERT INTO product (nickname, price, title) VALUES (%s, %s, %s)"
        
    cursor.execute(sql, (nickname, price, title))

db.commit()

f.close()
```

```python
f = open('product_info_data.csv', 'r', encoding='UTF-8')
csv_reader = csv.reader(f, delimiter=',')

print(csv_reader)


cnt = 0

cursor = db.cursor()

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
    
    category = (row[1])
    content = (row[2])
    delivery_fee = (row[3])
    info_notice = (row[4])
    reg_date = (row[5])
    stock = (row[6])
    upd_date = (row[7])
    product_id = (row[8])

        
    sql = "INSERT INTO product_info (category, content, delivery_fee, info_notice, reg_date, stock,upd_date, product_id) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
        
    cursor.execute(sql, (category, content,delivery_fee, info_notice, reg_date, stock, upd_date, product_id))

db.commit()

f.close()
```
